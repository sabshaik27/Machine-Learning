```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Poly_dataSet.csv')
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
sc_y = StandardScaler()
y = sc_y.fit_transform(np.reshape(y,(10,1)))
```

```python

# Fitting Linear Regression to the dataset
from sklearn.svm import SVR
lin_reg = SVR(kernel ='rbf')
lin_reg.fit(X, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (SVR Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
#Prdict by SV Regression
lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with SV Regression
predict = sc_y.inverse_transform(lin_reg.predict(sc_X.transform([[6.5]])).reshape(-1,1))
```
